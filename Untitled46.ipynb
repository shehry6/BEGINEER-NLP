```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences, to_categorical
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU, Dense, Embedding, Dropout, Bidirectional
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
stopwords = stopwords.words('english')
```

```python
# text preprocessing
# TEXT CLEANING FUNCTION
def text_cleaning(text):
    text = text.lower()                      # lower case
    text = re.sub(r'[^\w\s]', "",text )      # remove  punctuation
    text = re.sub(r"\d+","" , text)          # remove numbers
    text = ' '.join(word for word in text.split() if word not in stopwords) # remove stop words
    return text
```

```python
# APPLY TEXT CLEANING
data['text'] = data.text.apply(text_cleaning)
```

```python
data.head(3)
```

```python
tokenize = Tokenizer()
```

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size=0.2, random_state=42,shuffle=True,stratify=data['label'])
```
